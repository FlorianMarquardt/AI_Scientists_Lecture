## Some imports

````python
from experiment import run, solve # will require an installed version of sciexplorer
from matplotlib import pyplot as plt
import jax.numpy as jnp
%matplotlib inline
%config InlineBackend.figure_format='retina'
````

## Exploration

Run an experiment for given initial conditions,
q_init and q_dot_init (position and velocity).

````python
ts,Q=run(q_init=2.0, q_dot_init=-.01)

plt.plot(ts,Q[:,0])
plt.show()
````

Now simulate a hypothetical model that might reproduce the observed behaviour!

````python
# define the right-hand-side of a diff. equation
# (hint: params is not used, Q is an array of length 2)
def rhs(Q,t,params):
    return jnp.array([
        Q[1], # dQ[0]/dt
        -Q[0] # dQ[1]/dt
    ])

# initial conditions
Q0=jnp.array([2.,0.])

# solve the diff. equation (for standard duration and timestep)
Qsim=solve(Q0, rhs)

# plot the simulated trajectory
plt.plot(ts,Qsim[:,0])
plt.show()
````
